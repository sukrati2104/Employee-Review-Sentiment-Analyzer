```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df=pd.read_csv("Exploration.csv")
```

```python
df['Overall-Ratings']=np.where(df['Overall-Ratings']=='none',0,df['Overall-Ratings'])
df['Culture-Values']=np.where(df['Culture-Values']=='none',0,df['Culture-Values'])
df['Career-Opportunities'] =np.where(df['Career-Opportunities']=='none',0, df['Career-Opportunities'])
df['Company-Benefit'] = np.where(df['Company-Benefit'] =='none',0,df['Company-Benefit'])
df['Senior-Mangement'] = np.where(df['Senior-Mangement']=='none',0,df['Senior-Mangement'])
```

```python
df['Overall-Ratings']=pd.to_numeric(df['Overall-Ratings'])
df['Culture-Values']=pd.to_numeric(df['Culture-Values'])
df['Culture-Values'] = pd.to_numeric(df['Culture-Values'])
df['Career-Opportunities'] = pd.to_numeric(df['Career-Opportunities'])
df['Company-Benefit'] = pd.to_numeric(df['Company-Benefit'])
df['Senior-Mangement'] = pd.to_numeric(df['Senior-Mangement'])
```

It filters Google's review ratings for different categories (e.g., work balance, culture).

Converts review dates to year format.

Extracts the years of reviews for Google to analyze trends over time.

```python
x1=df['Overall-Ratings'][df['Company']=='google']
x2 = df['Work-Balance'][df['Company'] == 'google']
x3 = df['Culture-Values'][df['Company'] == 'google']
x4 = df['Career-Opportunities'][df['Company'] == 'google']
x5 = df['Company-Benefit'][df['Company'] == 'google']
x6 = df['Senior-Mangement'][df['Company'] == 'google']

df['Dates']=pd.to_datetime(df['Dates'],errors='coerce')
df['year']=df['Dates'].dt.year

y1=df['year'][df['Company']=='google']
```

Older years might be assigned lighter shades (like light blue).

More recent years might be assigned darker shades (like dark blue).

```python
# cheking google's Time-series

plt.rcParams['figure.figsize'] = (15, 15)

plt.subplot(3, 2, 1)
sns.boxenplot(x=y1, y=x1, palette = 'Blues', hue=y1, legend=False)
plt.title('Overall-rating')

plt.subplot(3, 2, 2)
sns.boxenplot(x=y1,y= x2, palette = 'Oranges', hue=y1, legend=False)
plt.title('work-balance-rating')

plt.subplot(3, 2, 3)
sns.boxenplot(x=y1, y=x3, palette = 'Purples', hue=y1, legend=False)
plt.title('Overall-rating')

plt.subplot(3, 2, 4)
sns.boxenplot(x=y1, y=x4, palette = 'Greys', hue=y1, legend=False)
plt.title('career-oppr-values-rating')

plt.subplot(3, 2, 5)
sns.boxenplot(x=y1, y=x5, palette = 'Greens', hue=y1, legend=False)
plt.title('benefit-rating')

plt.subplot(3, 2, 6)
sns.boxenplot(x=y1, y=x6, palette = 'Reds', hue=y1, legend=False)
plt.title('management-rating')
plt.show()
```

If a company has a wide box for Work Balance, it means:

Employee ratings vary widely.

Some think its great, others not so much.

If another has a narrow box:

Most employees agree on the rating, even if it's high or low.
